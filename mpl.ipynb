importing matplotlib library

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Values of x using numpy 

```python
x = np.arange(0,math.pi*2,0.05)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
print(z**2 + y**2)
```

Plotting Data

```python
#making a sin wave
plt.plot(x,y)
plt.show()
```

```python
#making cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x , y axis labels
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("sine wave")
plt.show()
```

```python
plt.plot(x,y, color='r', label='sine')
plt.plot(x,z, color='b',label='cosine')
plt.xlabel("angle")
plt.ylabel("sin(x),cos(x)")
plt.title("Sine and Cosine wave")
plt.legend()
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y,'g.')
```

```python
plt.plot(x,y,'bx')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English', 'Spanish','French','German']
people=[100,150,120,80]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No of People')
plt.show
```

Making a Pie-Chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English', 'Spanish','French','German']
people=[100,150,120,80]
ax.pie(people, labels=languages, autopct="%1.1f%%")
plt.show()
```

```python
#ScatterPlot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scatter Plot
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z= 20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z, c=z,cmap="Blues")
plt.show()
```
